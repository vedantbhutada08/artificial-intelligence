write a program to implement tic tac toe game .this problem is a generalized problem of'x or o' . represent the problem by using 3x3 matrix.player1 is a user and player 2 is a computer.automatic game playing will be stopped when either any one wins the game or game is drawn.computer i.e player 2 should generate randomly .also put condition of draw.provide a dotted line format for placing the x and o

```python
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-----------")

def is_winner(board, player):
    # Check rows, columns, and diagonals
    for i in range(3):
        if all(cell == player for cell in board[i]) or all(board[j][i] == player for j in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

def is_board_full(board):
    return all(all(cell != ' ' for cell in row) for row in board)

def user_move(board):
    while True:
        try:
            position = int(input("Enter your move (1-9): "))
            if 1 <= position <= 9:
                row, col = divmod(position - 1, 3)
                if board[row][col] == ' ':
                    board[row][col] = 'X'
                    break
                else:
                    print("Cell already taken. Try again.")
            else:
                print("Invalid position. Please enter a number between 1 and 9.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

def computer_move(board):
    while True:
        row = random.randint(0, 2)
        col = random.randint(0, 2)
        if board[row][col] == ' ':
            board[row][col] = 'O'
            break

def main():
    board = [[' ' for _ in range(3)] for _ in range(3)]

    while True:
        print_board(board)

        # Player 1 (User) move
        user_move(board)

        if is_winner(board, 'X'):
            print_board(board)
            print("Congratulations! Player 1 wins!")
            break

        if is_board_full(board):
            print_board(board)
            print("It's a draw!")
            break

        # Player 2 (Computer) move
        print("Player 2 is making a move...")
        computer_move(board)

        if is_winner(board, 'O'):
            print_board(board)
            print("Player 2 (Computer) wins! Better luck next time.")
            break

        if is_board_full(board):
            print_board(board)
            print("It's a draw!")
            break

if __name__ == "__main__":
    main()
```
